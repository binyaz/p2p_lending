```python
df_07_14 = df_14.append(df_12_13).append(df_07_11)
```

```python
df_07_14 = df_07_14[(df_07_14['loan_status'] == 'Charged Off')|(df_07_14['loan_status'] == 'Fully Paid')]
```

```python
df_07_14.shape
```

```python
df_07_14.isna().sum()
```

```python
df_07_14.dropna(axis = 1, how = 'all', inplace=True)
```

```python
df_07_14.drop(['hardship_flag', 'application_type', 'policy_code', 'desc', 'url', 'pymnt_plan', \
               'out_prncp', 'out_prncp_inv'], axis=1, inplace=True)
```

```python
df_07_14.info()
```

```python
df_07_14['loan_status'] = df_07_14['loan_status'].map({'Charged Off':0, 'Fully Paid':1})
df_07_14['grade'] = df_07_14['grade'].map({'A':1, 'B':2, 'C':3, 'D':4, 'E':5, 'F':6, 'G':7})
df_07_14['sub_grade'] = df_07_14['sub_grade'].map({'A1':11, 'A2':12, 'A3':13, 'A4':14, 'A5':15, \
                                                   'B1':21, 'B2':22, 'B3':23, 'B4':24, 'B5':25, \
                                                   'C1':31, 'C2':32, 'C3':33, 'C4':34, 'C5':35, \
                                                   'D1':41, 'D2':42, 'D3':43, 'D4':44, 'D5':45, \
                                                   'E1':51, 'E2':52, 'E3':53, 'E4':54, 'E5':55, \
                                                   'F1':61, 'F2':62, 'F3':63, 'F4':64, 'F5':65, \
                                                   'G1':71, 'G2':72, 'G3':73, 'G4':74, 'G5':75})
df_07_14['term'] = df_07_14['term'].map({' 36 months':36, ' 60 months':60})
df_07_14['emp_length'] = df_07_14['emp_length'].map({'< 1 years':-1, '1 years':1, '2 years':2, '3 year':3, \
                                                     '4 years':4, '5 years':5, '6 years':6, '7 years':7, \
                                                     '8 years':8, '9 year':9, '10+ years':10})
df_07_14['home_ownership'] = df_07_14['home_ownership'].map({'MORTGAGE':1, 'RENT':2, 'OWN':3, \
                                                             'ANY':4, 'NONE':5, 'OTHER':6})
df_07_14['verification_status'] = df_07_14['verification_status'].map({'Not Verified':0, 'Source Verified':2, \
                                                                       'Verified':1})
df_07_14['purpose'] = df_07_14['purpose'].map({'credit_card':1, 'debt_consolidation':2, 'car':3, 'house':4, \
                                               'home_improvement':5, 'other':6, 'medical':7, 'moving':8, \
                                               'major_purchase':9, 'vacation':10, 'small_business':11, \
                                               'renewable_energy':12, 'wedding':13, 'educational':14})
df_07_14['initial_list_status'] = df_07_14['initial_list_status'].map({'w':1, 'f':2})
df_07_14['debt_settlement_flag'] = df_07_14['debt_settlement_flag'].map({'N':0, 'Y':1})
```

```python
df_07_14['zip_code'] = df_07_14['zip_code'].map(lambda x: x[:3]).astype(int)
```

```python
df_07_14['issue_d'] = pd.to_datetime(df_07_14['issue_d'])
df_07_14['earliest_cr_line'] = pd.to_datetime(df_07_14['earliest_cr_line'])
df_07_14['last_pymnt_d'] = pd.to_datetime(df_07_14['last_pymnt_d'])
df_07_14['last_credit_pull_d'] = pd.to_datetime(df_07_14['last_credit_pull_d'])
df_07_14['hardship_start_date'] = pd.to_datetime(df_07_14['hardship_start_date'])
df_07_14['hardship_end_date'] = pd.to_datetime(df_07_14['hardship_end_date'])
df_07_14['payment_plan_start_date'] = pd.to_datetime(df_07_14['payment_plan_start_date'])
df_07_14['debt_settlement_flag_date'] = pd.to_datetime(df_07_14['debt_settlement_flag_date'])
df_07_14['settlement_date'] = pd.to_datetime(df_07_14['settlement_date'])
```

```python
df_07_14['int_rate'] = df_07_14['int_rate'].apply(lambda x:x.strip('%')).astype(float)
```

```python
df_07_14['id'] = df_07_14['id'].astype(int)
```
